```python
def delete_brand(df, brand):
    # Get names of indexes for which column value is specified
    indexNames = df[df['Brand'] == brand].index
    # Delete these row indexes from dataFrame
    df.drop(indexNames , inplace=True)
    #return df
```

```python
drop_brands = [
    'Micro-Swiss',
    'Creality',
    'Magigoo',
    'BCN3D',
    'Mosaic', 
    'Capricorn',
    'Polaroid3D'
]
```

```python
for brand in drop_brands:
    delete_brand(filament_df, brand)
```

```python
def drop_col_by_rowValue(df, col, val):
    if col in df.columns:
        # Get names of indexes for which column value is specified
        indexNames = df[df[col] == val].index
        # Delete these row indexes from dataFrame
        df.drop(indexNames , inplace=True)
        #return df
```

```python
drop_col_by_rowValue(filament_df, 'EAN', '6938936705246')
```

```python
drop_EANs = [
    '6938936705246', # Polymaker Polysher 
    '6938936701804', # Polymaker PolyBox Edition II
    '8435439882019', # bq Witbox Go! 
    '5902280822882', # Zortrax M200 Plus
    '5902280823704', # Zortrax M300 Plus
    '5902280823575', # Zortrax Filamentsensor + Kabel 
    'nan'
]
```

```python
for ean in drop_EANs:
    drop_col_by_rowValue(filament_df, 'EAN', ean)
```
