```python
def delete_brand(df, brand):
    # Get names of indexes for which column value is specified
    indexNames = df[df['Brand'] == brand].index
    # Delete these row indexes from dataFrame
    df.drop(indexNames , inplace=True)
    #return df
```

```python
def drop_col_by_rowValue(df, col, val):
    if col in df.columns:
        # Get names of indexes for which column value is specified
        indexNames = df[df[col] == val].index
        # Delete these row indexes from dataFrame
        df.drop(indexNames , inplace=True)
        #return df
```

## Extracting 'Weight' information

```python
bq_df = filament_df[filament_df.Brand =='bq']
bq_df
```

```python
def bq_colour(df):
    try:
        colour_str = df['Title'].replace("bq Easy Go PLA ", "")
        colour = colour_str.split('-')[0]
    except: 
        colour = df['Title']
    return colour
```

```python
def bq_weight(df):
    if df['Title'].find('/'):
        try:
            weight = df['Title'].split('/')[-1]
            weight = weight.split()[0]
        except:
            weight = ""
    else:
        weight = ""
    return weight
```

```python
filament_df['Material'][filament_df['Title'].str.startswith("bq Easy Go PLA")] = "PLA"
filament_df['Product_line'][filament_df['Title'].str.startswith("bq Easy Go PLA")] = "bq Easy Go PLA"
filament_df['Colour'][filament_df['Title'].str.startswith("bq Easy Go PLA")] = filament_df.apply(lambda x: bq_colour(x), axis=1)
filament_df['Weight'][filament_df['Title'].str.startswith("bq Easy Go PLA")] = filament_df.apply(lambda x: bq_weight(x), axis=1)
```
