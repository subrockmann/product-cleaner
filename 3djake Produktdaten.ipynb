## Create sub dataframe only with filaments

```python
filament_df = df[df['Description'].str.contains('Filament')].copy()
```

```python
filament_df['Diameter']= ''
```

```python
filament_df.loc[filament_df['Title'].str.contains("1,75"), 'Diameter'] = '1.75'
filament_df.loc[filament_df['Description'].str.contains("Zortrax M200"), 'Diameter'] = '1.75'
filament_df.loc[filament_df['Title'].str.contains("Z-ULTRAT"), 'Diameter'] = '1.75'
filament_df.loc[filament_df['Title'].str.contains("Z-HIPS"), 'Diameter'] = '1.75'
filament_df.loc[filament_df['Title'].str.contains("Z-PETG"), 'Diameter'] = '1.75'
filament_df.loc[filament_df['Title'].str.contains("Z-ESD"), 'Diameter'] = '1.75'

filament_df.loc[filament_df['Title'].str.contains("2,85"), 'Diameter'] = '2.85'
```
