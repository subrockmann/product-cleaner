```python
def bq_weight(df):
    if df['Title'].find('/'):
        try:
            weight = df['Title'].split('/')[-1]
            weight = weight.split()[0]
        except:
            weight = ""
    else:
        weight = ""
    return weight
```

```python
def bq_colour(df):
    try:
        colour_str = df['Title'].replace("bq Easy Go PLA ", "")
        colour = colour_str.split('-')[0]
    except: 
        colour = df['Title']
    return colour
```

```python
filament_df['Material'][filament_df['Title'].str.startswith("bq Easy Go PLA")] = "PLA"
filament_df['Product_line'][filament_df['Title'].str.startswith("bq Easy Go PLA")] = "bq Easy Go PLA"
filament_df['Colour'][filament_df['Title'].str.startswith("bq Easy Go PLA")] = filament_df.apply(lambda x: bq_colour(x), axis=1)
filament_df['Weight'][filament_df['Title'].str.startswith("bq Easy Go PLA")] = filament_df.apply(lambda x: bq_weight(x), axis=1)
```

```python
def extrudr_colour(df):
    try:
        colour_str = df['Title'].split('-') #-
        if len(colour_str) == 2:
            #colour_str = 1
            colour_str = colour_str[0]
        elif len(colour_str) == 3: 
            #colour_str = "3"
            colour_str = str(colour_str[0]) +'-' +str(colour_str[1])
            #print(colour_str)
        else:
            colour_str = ""
            
        product_line = df['Product_line']
        colour = colour_str.replace(product_line, "").strip()
        #colour = colour_str
    except: 
        colour = df['Title']
    return colour
```

```python
def zortrax_productline(df):
    if df['Title'].startswith("Zortrax Z-ABS"):
        product_line = "Zortrax Z-ABS"
        material = "ABS"
    elif df['Title'].startswith("Zortrax Z-ASA Pro"):
        product_line = "Zortrax Z-ASA Pro"
        material = "ASA"     
    elif df['Title'].startswith("Zortrax Z-ESD"):
        product_line = "Zortrax Z-ESD"
        material = "ESD"
    elif df['Title'].startswith("Zortrax Z-FLEX"):
        product_line = "Zortrax Z-FLEX"
        material = "Flex"
    elif df['Title'].startswith("Zortrax Z-HIPS"):
        product_line = "Zortrax Z-HIPS"
        material = "HIPS"   
    elif df['Title'].startswith("Zortrax Z-NYLON"):
        product_line = "Zortrax Z-NYLON"
        material = "Nylon" 
    elif df['Title'].startswith("Zortrax Z-PETG"):
        product_line = "Zortrax Z-PETG"
        material = "PETG"
    elif df['Title'].startswith("Zortrax Z-PLA"):
        product_line = "Zortrax Z-PLA"
        material = "PLA"
    elif df['Title'].startswith("Zortrax Z-SEMIFLEX"):
        product_line = "Zortrax Z-SEMIFLEX"
        material = "Flex"
    elif df['Title'].startswith("Zortrax Z-ULTRAT"):
        product_line = "Zortrax Z-ULTRAT"
        material = "?"
        
    else:
        product_line = ""
        material = ""
    return product_line, material
```

```python
def zortrax_weight(df):
    if df['Title'].find('-'):
        try:
            weight = df['Title'].split('-')[-1]
            weight = weight.split()[0]
        except:
            weight = ""
    else:
        weight = ""
    return weight
```
